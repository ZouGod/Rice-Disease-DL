```python
import os
import shutil

# Define the source directories
source_dir_filed = r'D:\CADT\CapstoneProjectI\ml__model\data\balance_data\field'
source_dir_white = r'D:\CADT\CapstoneProjectI\ml__model\data\balance_data\white'

# Define the destination directory
destination_dir = r'D:\CADT\CapstoneProjectI\ml__model\data\raw_images'

# Create the destination directory if it doesn't exist
os.makedirs(destination_dir, exist_ok=True)

# Function to copy files from one directory to another
def copy_files(source_dir, destination_dir):
    for folder in os.listdir(source_dir):
        folder_path = os.path.join(source_dir, folder)
        if os.path.isdir(folder_path):
            # Create the same folder structure in the destination
            dest_folder_path = os.path.join(destination_dir, folder)
            os.makedirs(dest_folder_path, exist_ok=True)
            # Copy files from source to destination
            for file in os.listdir(folder_path):
                file_path = os.path.join(folder_path, file)
                if os.path.isfile(file_path):
                    shutil.copy(file_path, dest_folder_path)

# Copy files from both source directories
copy_files(source_dir_filed, destination_dir)
copy_files(source_dir_white, destination_dir)

print("Files have been successfully combined into:", destination_dir)
```

```python
import os
import matplotlib.pyplot as plt

# Define the combined directory
combined_dir = r'D:\CADT\CapstoneProjectI\ml__model\data\Rice Diseases\field background'

# Initialize a dictionary to hold the counts
class_counts = {}

# Count images in each subfolder
for folder in os.listdir(combined_dir):
    folder_path = os.path.join(combined_dir, folder)
    if os.path.isdir(folder_path):
        # Count the number of image files (assuming .jpg and .png formats)
        image_count = len([file for file in os.listdir(folder_path) if file.endswith(('.jpg', '.png'))])
        class_counts[folder] = image_count

# Plotting
classes = list(class_counts.keys())
counts = list(class_counts.values())

plt.figure(figsize=(10, 6))
plt.bar(classes, counts, color='skyblue')
plt.xlabel('Classes')
plt.ylabel('Number of Images')
plt.title('Number of Images in Each Class')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import os
import matplotlib.pyplot as plt
import numpy as np

# Define the combined directory
combined_dir = r'D:\CADT\CapstoneProjectI\ml__model\data\balance_data\field'

# Initialize a dictionary to hold the counts
class_counts = {}

# Count images in each subfolder
for folder in os.listdir(combined_dir):
    folder_path = os.path.join(combined_dir, folder)
    if os.path.isdir(folder_path):
        # Count the number of image files (assuming .jpg and .png formats)
        image_count = len([file for file in os.listdir(folder_path) if file.endswith(('.jpg', '.png'))])
        class_counts[folder] = image_count

# Plotting
classes = list(class_counts.keys())
counts = list(class_counts.values())

# Generate a color map
colors = plt.cm.viridis(np.linspace(0, 1, len(classes)))

plt.figure(figsize=(10, 6))
bars = plt.bar(classes, counts, color=colors)
plt.xlabel('Classes')
plt.ylabel('Number of Images')
plt.title('Number of Images in Each Class Field Background')
plt.xticks(rotation=45)

# Display the count of images on top of each bar
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, yval, int(yval), ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

```python
import os
import cv2
import matplotlib.pyplot as plt

# Define the directory containing the images
combined_dir = r'D:\CADT\CapstoneProjectI\ml__model\data\raw_images'

# Class names based on your previous data
class_names = [
    "Bacterial Blight",
    "Brown Spot",
    "Healthy",
    "Hispa",
    "Leaf Blast",
    "Leaf Scald",
    "Leaf Smut",
    "Narrow Brown Leaf Spot",
    "Neck Blast",
    "Sheath Blight",
    "Tungro",
]

# Prepare a list to hold the images
images = []

# Load one image from each class
for class_name in class_names:
    class_path = os.path.join(combined_dir, class_name)
    if os.path.isdir(class_path):
        # Get the first image file in the directory
        image_files = [f for f in os.listdir(class_path) if f.endswith(('.jpg', '.png'))]
        if image_files:
            image_path = os.path.join(class_path, image_files[0])
            image = cv2.imread(image_path)
            if image is not None:
                # Convert BGR to RGB
                image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
                images.append(image)

# Plotting
plt.figure(figsize=(15, 8))
for i, (image, class_name) in enumerate(zip(images, class_names)):
    plt.subplot(3, 5, i + 1)  # 2 rows, 5 columns
    plt.imshow(image)
    plt.title(class_name)
    plt.axis('off')

plt.tight_layout()
plt.show()
```
